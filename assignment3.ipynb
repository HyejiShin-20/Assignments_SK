### 문제 1. 주문 항목 객체 만들기

카페 주문 시스템에서 “주문 항목”을 객체로 표현하려고 합니다.

아래 요구사항을 만족하는 `OrderItem` 클래스를 작성하세요.

1. 속성: `name`, `unit_price`, `qty`
2. 메소드:
    - `total_price()` : `unit_price * qty` 반환
    - `__str__()` : `"아메리카노 x2 = 6000원"` 형태로 반환
3. 실행 예시를 참고하여 출력이 나오도록 테스트 코드도 작성

기대 출력 예시
아메리카노 x2 = 6000원
치즈케이크 x1 = 5500원
총합: 11500원

```python
class OrderItem:
    def __init__(self, name, unit_price, qty):
        self.name = name
        self.unit_price = unit_price
        self.qty = qty
    def total_price(self):
        return self.unit_price * self.qty
    def __str__(self):
        return f"{self.name} x{self.qty} = {self.total_price()}원"
    
order1 = OrderItem("아메리카노", 3000, 2)
order2 = OrderItem("치즈케이크", 5500, 1)

print(order1)
print(order2)
print(f"총합: {order1.total_price() + order2.total_price()}원")
    
```

### 문제 3. 상속으로 메뉴 확장

`MenuItem`을 부모 클래스로 두고 `Coffee`, `Dessert`가 상속받도록 구성하세요.

1. `MenuItem` 공통 속성: `name`, `unit_price`
2. `Coffee`는 추가 속성: `shot` (기본 1)
3. `Dessert`는 추가 속성: `size` ("S", "M", "L" 중 하나)
4. `get_info()` 메서드:
    - Coffee: `"아메리카노(샷 2) - 3000원"`
    - Dessert: `"치즈케이크(L) - 5500원"`
5. 다형성: `items = [Coffee(...), Dessert(...)]` 같은 리스트에서 `for`로 `get_info()`를 호출해도 동일하게 동작

```python
class MenuItem:
    def __init__(self, name, unit_price):
        self.name = name
        self.unit_price = unit_price
        
class Coffee(MenuItem):
    def __init__(self, name, unit_price, shot):
        super().__init__(name, unit_price)
        self.shot = shot
    def get_info(self):
        if self.shot > 0:
            return f"{self.name}(샷 {self.shot}) - {self.unit_price}원"
        else:
            return f"{self.name} - {self.unit_price}원"

class Dessert(MenuItem):
    def __init__(self, name, unit_price, size):
        super().__init__(name, unit_price)
        if size == "S" or size == "M" or size == "L":
            self.size = size
        else:
            return "잘못된 사이즈입니다."
    def get_info(self):
        return f"{self.name}({self.size}) - {self.unit_price}원"
    
items = [Coffee("아메리카노", 3000, 2), Dessert("치즈케이크", 5500, "L")]
for MenuItem in items:
    print(MenuItem.get_info())
```

### 문제 4. 장바구니(Cart)로 집계하기

`Cart` 클래스를 만들고 아래 기능을 구현하세요.

1. `add(item: OrderItem)` : 장바구니에 담기
2. `total()` : 장바구니 총합 반환
3. `receipt()` : 영수증 출력 (각 항목 + 총합)

```python
class OrderItem:
    def __init__(self, name, price, qty):
        if qty > 0 :
            self.name = name
            self.price = price
            self.qty = qty
            self.itemSet = {"name": self.name, "price": self.price, "qty": self.qty}
        else:
            return "수량을 0 이상 기입해주세요."
    def __str__(self):
        return f"{self.name} : {self.price}원 [{self.qty}개]"
        
class Cart():
    def __init__(self, user, user_cart=[]):
        self.user = user
        self.user_cart = user_cart
        
    def add(self, item: OrderItem):
        self.user_cart.append(item.itemSet)
        print(f"{item.name} {item.qty}개를 장바구니에 추가했습니다.")
    
    def total(self):
        total = 0
        for elements in self.user_cart:
            total += elements['qty'] * elements['price']
        return total
    
    def receipt(self):
        print ("========  영수증  ========")
        for elements in self.user_cart:
            print(f"{elements['name']} ({elements['qty']}개) - {elements['qty'] * elements['price']}원")
        print(f"\n최종 금액 : {self.total()}원")
        
    def clear_cart(self):
        self.user_cart.clear()
        # print("장바구니를 비웠습니다.")

kim = Cart("kim")
kim.clear_cart
kim.add(OrderItem("아메리카노 원두", 15000, 10))
kim.add(OrderItem("핸드폰 충전기", 10000, 1))
kim.add(OrderItem("건전지", 18000, 3))

kim.receipt()
```

### 문제 5. 결제 전략(Strategy) 흉내내기

결제 수단을 객체로 분리해서 처리하세요.

1. `Payment` 추상 클래스(또는 단순 인터페이스 역할) 정의: `pay(amount)` 메소드
2. `CardPayment`, `CashPayment` 구현
    - Card: `"카드로 {amount}원 결제 완료"`
    - Cash: `"현금으로 {amount}원 결제 완료"`
3. `Cart.checkout(payment)` 호출 시 총합을 계산해 결제하고, 결제 메시지를 출력

```python
from abc import ABC, abstractmethod
class Cart:
    def checkout(self, method):
        method.pay(self.total())
        self.clear_cart()
        
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass
    
class CardPayment(Payment):
    def pay(self, amount): print(f"카드로 {amount}원 결제 완료")
class CashPayment(Payment):
    def pay(self, amount): print(f"현금으로 {amount}원 결제 완료")
    
kim.receipt()
Cart.checkout(kim, CardPayment())
```
