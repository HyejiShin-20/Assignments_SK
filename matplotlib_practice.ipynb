```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

## 공통 데이터 준비

아래 데이터는 카페 1개 지점의 최근 6개월 운영 데이터이다.
- sales: 월매출(만원)
- visitors: 방문 고객 수
- ad_spend: 광고비(만원)
- avg_spend: 1인 평균 결제 금액(원)

```python

months = ['1월', '2월', '3월', '4월', '5월', '6월']

cafe_df = pd.DataFrame({
    'month': months,
    'sales': [320, 340, 360, 390, 420, 460],
    'visitors': [1200, 1280, 1350, 1500, 1620, 1750],
    'ad_spend': [40, 45, 50, 60, 68, 75],
    'avg_spend': [8700, 8900, 9000, 9300, 9500, 9800]
})

cafe_df
```

## 문제 1. 월별 매출 추이 Line Plot

요구사항
1. x축은 month, y축은 sales로 설정한다.
2. 선 색상은 blue, 마커는 o로 설정한다.
3. 그래프 제목은 '월별 매출 추이'로 설정한다.
4. x축 이름은 '월', y축 이름은 '매출(만원)'으로 설정한다.
5. grid를 표시한다.

질문
- 매출이 가장 크게 증가한 구간은 어디인가?

```python

# TODO: 월별 매출 추이 line plot 완성하기
plt.figure(figsize=(8, 4))

X = cafe_df['month']
Y = cafe_df['sales']
line = plt.gca()
line.plot(X, Y, color='blue', marker='o')
plt.title('월별 매출 추이')
plt.xlabel('월')
plt.ylabel('매출(만원)')
plt.grid(True)
plt.show()

# 매출이 가장 크게 증가한 구간 : 5월 -> 6월
```

## 문제 2. 월별 방문자 수 Bar Plot

요구사항
1. x축은 month, y축은 visitors로 설정한다.
2. 막대 색상은 orange로 설정한다.
3. 그래프 제목은 '월별 방문자 수'로 설정한다.
4. x축 이름과 y축 이름을 알맞게 설정한다.

```python

# TODO: 월별 방문자 수 bar plot 완성하기

plt.figure(figsize=(8, 4))

ax = plt.gca()
X = cafe_df['month']
Y = cafe_df['visitors']
ax.bar(X, Y, color='orange')
plt.xlabel("월")
plt.ylabel("방문자 수")
plt.title("월별 방문자 수")
plt.show()
```

## 문제 3. 일별 결제 금액 분포 Histogram

카페에서는 일별 결제 금액이 어느 구간에 많이 몰려 있는지도 보고 싶어 한다.

요구사항
1. 아래 daily_pay 배열을 사용한다.
2. bins=8로 설정한다.
3. edgecolor='black'을 설정한다.
4. 그래프 제목은 '일별 결제 금액 분포'로 설정한다.

```python

daily_pay = [4100, 5300, 6200, 4800, 7100, 8300, 9200, 10400, 11600, 9800,
             8700, 7600, 6700, 5900, 12500, 13800, 14900, 9100, 8400, 7900]

# TODO: histogram 완성하기

plt.figure(figsize=(8, 4))

plt.hist(
    daily_pay,
    bins=8,
    edgecolor='black'
)
plt.title("일별 결제 금액 분포")
plt.show()
```

## 문제 4. 광고비와 매출의 관계 Scatter Plot

마케팅팀은 광고비가 늘면 매출도 함께 늘어나는지 보고 싶어 한다.

요구사항
1. x축은 ad_spend, y축은 sales로 설정한다.
2. 점 색상은 green으로 설정한다.
3. 그래프 제목은 '광고비와 매출의 관계'로 설정한다.
4. x축 이름은 '광고비(만원)', y축 이름은 '매출(만원)'으로 설정한다.

질문
- 점들의 분포를 봤을 때 양의 관계가 있는가?

```python

# TODO: scatter plot 완성하기

plt.figure(figsize=(6, 4))

plt.scatter(
    x=cafe_df['ad_spend'],
    y=cafe_df['sales'],
    colorizer='green'
)
plt.xlabel('광고비(만원)')
plt.ylabel('매출(만원)')
plt.title('광고비와 매출의 관계')

plt.show()
```

## 문제 5. 두 개의 지표를 한 화면에 비교하기

한 Figure 안에 2개의 Axes를 만들어 아래 그래프를 동시에 출력하라.

왼쪽 그래프
- month / sales line plot

오른쪽 그래프
- month / visitors bar plot

추가 조건
- figsize=(12, 4)
- 각 그래프에 title을 붙인다.

```python

# TODO: subplots를 이용해 2개 그래프를 한 화면에 배치하기

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(cafe_df['month'], cafe_df['sales'])
ax[0].set_title('Month / Sales line plot')
ax[1].bar(cafe_df['month'], cafe_df['visitors'])
ax[1].set_title('Month / visitors bar plot')
plt.tight_layout()
plt.show()
```

## 문제 6. 요일별 주문 건수 Box Plot

점심/저녁 시간대 운영 차이를 보기 전에, 먼저 요일별 주문 건수 분포를 박스플롯으로 확인하려고 한다.

요구사항
1. 아래 orders_by_day 데이터를 사용한다.
2. plt.boxplot()을 사용한다.
3. x축 눈금 이름을 월~일로 설정한다.
4. 제목은 '요일별 주문 건수 분포'로 설정한다.

```python

orders_by_day = [
    [42, 45, 39, 47, 50, 44],   # 월
    [40, 41, 43, 45, 46, 44],   # 화
    [44, 46, 48, 45, 47, 49],   # 수
    [43, 42, 44, 46, 45, 47],   # 목
    [50, 52, 49, 55, 58, 54],   # 금
    [60, 65, 68, 70, 66, 72],   # 토
    [55, 58, 60, 57, 62, 59]    # 일
]

# TODO: box plot 완성하기

plt.figure(figsize=(9, 4))

# 여기에 작성
plt.boxplot(
    orders_by_day,
    notch=True
)
plt.xlabel('월~일')
plt.title('요일별 주문 건수 분포')

plt.show()
```
