### 문제 1. 안전한 정수 입력 함수

사용자에게 숫자를 입력받는 함수 `input_int(prompt)`를 작성하세요.

1. 정수가 아니면 `"정수를 입력하세요."`를 출력하세요.
2. 올바른 정수를 입력할 때까지 반복하세요.
3. 정상 입력 시 `int` 값을 반환하세요.

```python
def input_int(prompt):
    while True:
        try:
            num = int(input(prompt))
            if num:
                print(num)
                return num
        except ValueError:
            print("정수를 입력하세요.")
get_num = input_int("숫자를 입력하세요: ")
```

### 문제 2. 나눗셈 계산기

두 정수를 입력받아 `a / b` 결과를 출력하세요.

1. 0으로 나누면 `"0으로 나눌 수 없습니다."`를 출력하세요.
2. 정수가 아닌 입력이면 `"정수만 입력하세요."`를 출력하세요.
3. 어떤 경우든 프로그램이 종료되지 않고 다시 입력받도록 작성하세요.

```python
def division_two_values(prompt):
    while True:
        try:
            a, b = int(input(prompt).split())
            total = a / b
            if total:
                print(f"{a} / {b} = {total}")
                return total
        except TypeError:
            print("정수만 입력하세요.")
        # except ValueError:
        #     print("정수 두 개를 입력하세요.")
        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다.")
            
result = division_two_values("두 숫자를 입력해 주세요: ")
```

### 문제 4. 사용자 정의 예외로 규칙 강제

점수 입력 규칙: **0~100 사이 정수만 허용**

1. 규칙을 위반하면 `InvalidScoreError` 예외를 발생시키세요.
2. main 로직에서 해당 예외를 잡아 `"점수는 0~100 사이여야 합니다."`를 출력하세요.
3. 올바른 점수를 입력할 때까지 반복하세요.

```python
class InvalidScoreError(Exception):
    ''' 점수 범위 0 ~ 100 '''
    
def validate(score):
    if not isinstance(score, (int, float)):
        raise TypeError("점수는 숫자여야 합니다.")
    if not (0 <= score <= 100):
        raise InvalidScoreError("점수는 0~100 사이여야 합니다.")
    
while True:
    try:
        score = int(input())
        validate(score)
        if score:
            print(f"{score}점을 입력했습니다.")
            break
    except InvalidScoreError:
        print ("점수는 0~100 사이여야 합니다.")
    except TypeError:
        print ("점수는 숫자여야 합니다.")
    except ValueError:
        print ("점수는 숫자여야 합니다.")
    
```

### 문제 5. finally 사용하기

프로그램 실행 중 예외가 발생할 수 있는 코드를 작성하세요.

조건

1. 예외가 발생하면 `"오류가 발생했습니다."`를 출력하세요.
2. 어떤 경우든 `"프로그램을 종료합니다."`가 마지막에 출력되도록 `finally`를 사용하세요.

```python
class InvalidScoreError(Exception):
    ''' 점수 범위 0 ~ 100 '''
    
def validate(score):
    if not isinstance(score, (int, float)):
        raise TypeError("점수는 숫자여야 합니다.")
    if not (0 <= score <= 100):
        raise InvalidScoreError("점수는 0~100 사이여야 합니다.")
    
while True:
    try:
        score = int(input())
        validate(score)
        if score:
            print(f"{score}점을 입력했습니다.")
            break
    except:
        print("오류가 발생했습니다.")
    finally:
        print("프로그램을 종료합니다.")
        
```
